# Determining Unloaded Quality Factor from a Two Port Network
Direct measurement of the unloaded Quality Factor $Q_0$ of a resonator is generally not possible because of the loading effect of the measurement system [1]. However, it is possible to determine $Q_0$ from measurements of the frequency response of the loaded resonator when it is connected to a transmission line. The examples below describe how unloaded $Q$ can be found from a two-port measurement using the `Qfactor` class of `scikit-rf`.

```python
import skrf as rf
import numpy as np
rf.stylely()
```

## Example 1: Series RLC
This example is inspired from [1, sec. 6.6]. Assume a series RLC resonator inserted in series in a transmission line of characteristic impedance $Z_0$ and wavenumber $\gamma$ as illustrated below: 

<img src="figures/twoport_RLC_series.svg" width="400"/>

Maximum transmission

```python
R = 0.0005  # Ohm
L = 1e-9  # H
C = 1e-12  # F
freq = rf.Frequency(1, 10, npoints=1001, unit='GHz')
media = rf.DefinedGammaZ0(freq, z0=50)
res = media.resistor(R)
ind = media.inductor(L)
cap = media.capacitor(C)
RLC = res ** ind ** cap

RLC.plot_s_db(m=0, n=0, lw=2)  # S11 Magnitude in dB
```

The theorical values of the resonance frequency and Q-factor are known in this case: 
$$ f_0 = \frac{1}{2 \pi \sqrt{L C} }  $$
$$ Q_0 = \frac{1}{R} \sqrt{\frac{L}{C}} $$



```python
f_th = 1/(2*np.pi*np.sqrt(L*C))
Q_th = np.sqrt(L/C)

print(f'Theoertical Resonance Frequency: {f_th/1e9} GHz')
print(f'Theoretical Quality Factor: {Q_th}')
```

```python
RLC.s11.plot_s_db()
```

```python
QF = rf.Qfactor(RLC.s11, res_type='reflection', verbose=True)
```

```python
QF.Q_L
```

```python
QF.f_L
```

```python
RLC.plot_s_polar(m=0, n=0)
RLC['5GHz'].plot_s_polar(m=0, n=0, marker='o')
```

```python
res = QF.fit(method='NLQFIT6')
print(res)
```

```python
QF.Q_unloaded(opt_res=res, A='auto')
```

```python
import tkinter
from tkinter import ttk
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox
from ttkthemes import ThemedTk
import os
import openpyxl

def openNewWindow():
     
    # Toplevel object which will 
    # be treated as a new window
    newWindow = Toplevel(window)
 
    # sets the title of the
    # Toplevel widget
    newWindow.title("Instructions")
 
    # sets the geometry of toplevel
    newWindow.geometry("600x200")
 
    # A Label widget to show in toplevel
    Label(newWindow, 
          text ="Fill in the keywords associated with each section.").pack()
    Label(newWindow, 
          text ="Keywords can include singular words or can be groups/phrases.").pack()
    Label(newWindow, 
          text ="Keywords entered must be separated by a common and trailing space ex: ', '.").pack()
    Label(newWindow, 
          text ="Example: chest pain, dull, short of breath, troponin").pack()
    
def enter_data():
    accepted = accept_var.get()
    
    if accepted=="Accepted":
        # case info
        case = case_name_entry.get()
        system = system_combobox.get()
        icd10 = icd_name_entry.get()
        
        if case and icd10 and system:
            hpi = history_hpi_entry.get()
            cc = history_cc_entry.get()
            ros = history_ros_entry.get()
            # Course info
            
            
            obj_vit=obj_vit_entry.get()
            obj_gen=obj_gen_entry.get()
            obj_dl=obj_dl_entry.get()
            obj_di=obj_di_entry.get()
            
            assess_dx=assess_dx_entry.get()
            assess_ddx=assess_ddx_entry.get()
            
            plan_tx=plan_tx_entry.get()
            plan_con=plan_con_entry.get()
            plan_inter=plan_inter_entry.get()
            
            found = found_entry.get()
            
            
            print("Case name: ", case, "ICD 10: ",icd10,"System name: ", system)
            print("History CC: ", cc, "History HPI: ", hpi, "History ROS: ", ros)
            print("Obj. Vitals: ", obj_vit, "Obj. GEN: ",obj_gen,"Obj. DL: ",obj_dl,"Obj. DI: ",obj_di )
            print("Assessment Dx: ", assess_dx,"Assessment DDx: ", assess_ddx)
            print("Plan Tx: ", plan_tx,"Plan Consults: ", plan_con,"Plan Interventions: ",plan_inter)
            print("Foundations: ", found)
            print("------------------------------------------")
            
            filepath = "Case Presentation KeyBank - Transpose - Blank.xlsx"
            
            if not os.path.exists(filepath):
                workbook = openpyxl.Workbook()
                sheet = workbook.active
                heading = ["Case", "ICD10","System", "History - CC","History - HPI", "History - ROS",
                            "History - MEDS", "Objective - VITALS","Objective - GEN","Objective - PE", "Objective - DL",
                             "Objective - DI", "Assessment - Dx", "Assessment - DDx", "Plan - Tx",
                              "Plan - Consult", "Plan - Interventions","Foundation"]
                sheet.append(heading)
                workbook.save(filepath)
            workbook = openpyxl.load_workbook(filepath)
            sheet = workbook.active
            count=0                        
            for row in sheet.iter_rows(min_row=1, max_col=2):
                for cell in row:
                    if cell.value != icd10:
                        count+=0 
                    else:
                        count =1
                        break
                        #workbook.save(filepath)
                        #workbook.close()
            if count == 0: 
                sheet.append([case,icd10,system,cc,hpi,ros,
                            obj_vit,obj_gen,obj_dl,obj_di,
                            assess_dx,assess_ddx,
                            plan_tx,plan_con,plan_inter,
                            found])    
            else:
                tkinter.messagebox.showwarning(title="Error", message="Case already in system.")
            workbook.save(filepath)
                
        else:
            tkinter.messagebox.showwarning(title="Error", message="Case, System, and ICD 10 are required.")
    else:
        tkinter.messagebox.showwarning(title= "Error", message="You have not accepted the terms")

#window = tkinter.Tk()
window = ThemedTk(theme="plastik")
#window.set_theme("dark")
#master.geometry("200x200")
window.title("Data Entry Form")
label = Label(window, 
              text ="This is the main window")
 
label.pack(pady = 10)
 
# a button widget which will open a 
# new window on button click
btn = Button(window, 
             text ="Click to view Instructions", 
             command = openNewWindow)
btn.pack(pady = 10)

frame = tkinter.Frame(window)
frame.pack()

# Saving case Info
case_info_frame =tkinter.LabelFrame(frame, text="Case Information")
case_info_frame.grid(row= 0, column=0, padx=20, pady=10)

case_name_label = tkinter.Label(case_info_frame, text="Case Name")
case_name_label.grid(row=0, column=0)
case_name_entry = tkinter.Entry(case_info_frame)
case_name_entry.grid(row=1, column=0)

icd_name_label = tkinter.Label(case_info_frame, text="ICD 10")
icd_name_label.grid(row=2, column=0)
icd_name_entry = tkinter.Entry(case_info_frame)
icd_name_entry.grid(row=3, column=0)

system_name_label = tkinter.Label(case_info_frame, text="System Name")
system_combobox = ttk.Combobox(case_info_frame, values=["MSK", "Neuro", "Cardiopulmonary", "GI","GU"])
system_name_label.grid(row=0, column=1)
system_combobox.grid(row=1, column=1)


for widget in case_info_frame.winfo_children():
    widget.grid_configure(padx=10, pady=5)
#########################################################################

history_info_frame =tkinter.LabelFrame(frame, text="History Information")
history_info_frame.grid(row= 1, column=0, padx=20, pady=10)

history_cc_label = tkinter.Label(history_info_frame, text="History - CC")
history_cc_label.grid(row=2, column=0)
history_cc_entry = tkinter.Entry(history_info_frame)
history_cc_entry.grid(row=3, column=0)


history_hpi_label = tkinter.Label(history_info_frame, text="History - HPI")
history_hpi_label.grid(row=2, column=1)
history_hpi_entry = tkinter.Entry(history_info_frame)
history_hpi_entry.grid(row=3, column=1)

history_ros_label = tkinter.Label(history_info_frame, text="History - ROS")
history_ros_label.grid(row=2, column=1)
history_ros_entry = tkinter.Entry(history_info_frame)
history_ros_entry.grid(row=3, column=1)

for widget in history_info_frame.winfo_children():
    widget.grid_configure(padx=10, pady=5)
###########################################################################
objective_info_frame =tkinter.LabelFrame(frame, text="Objective Information")
objective_info_frame.grid(row= 2, column=0, padx=20, pady=10)

obj_vit_label = tkinter.Label(objective_info_frame, text="Objective - VITALS")
obj_vit_label.grid(row=2, column=0)
obj_vit_entry = tkinter.Entry(objective_info_frame)
obj_vit_entry.grid(row=3, column=0)

obj_gen_label = tkinter.Label(objective_info_frame, text="Objective - GEN")
obj_gen_label.grid(row=2, column=1)
obj_gen_entry = tkinter.Entry(objective_info_frame)
obj_gen_entry.grid(row=3, column=1)

obj_dl_label = tkinter.Label(objective_info_frame, text="Objective - DL")
obj_dl_label.grid(row=4, column=0)
obj_dl_entry = tkinter.Entry(objective_info_frame)
obj_dl_entry.grid(row=5, column=0)

obj_di_label = tkinter.Label(objective_info_frame, text="Objective - DI")
obj_di_label.grid(row=4, column=1)
obj_di_entry = tkinter.Entry(objective_info_frame)
obj_di_entry.grid(row=5, column=1)
for widget in objective_info_frame.winfo_children():
    widget.grid_configure(padx=10, pady=5)
##########################################################################
assess_info_frame =tkinter.LabelFrame(frame, text="Assessment Information")
assess_info_frame.grid(row= 3, column=0, padx=20, pady=10)

assess_dx_label = tkinter.Label(assess_info_frame, text="Assessment - Dx")
assess_dx_label.grid(row=2, column=0)
assess_dx_entry = tkinter.Entry(assess_info_frame)
assess_dx_entry.grid(row=3, column=0)


assess_ddx_label = tkinter.Label(assess_info_frame, text="Assessment - DDx")
assess_ddx_label.grid(row=2, column=1)
assess_ddx_entry = tkinter.Entry(assess_info_frame)
assess_ddx_entry.grid(row=3, column=1)

for widget in assess_info_frame.winfo_children():
    widget.grid_configure(padx=10, pady=5)
#########################################################################

plan_info_frame =tkinter.LabelFrame(frame, text="Plan Information")
plan_info_frame.grid(row= 4, column=0, padx=20, pady=10)

plan_tx_label = tkinter.Label(plan_info_frame, text="Plan - Tx")
plan_tx_label.grid(row=2, column=0)
plan_tx_entry = tkinter.Entry(plan_info_frame)
plan_tx_entry.grid(row=3, column=0)

plan_con_label = tkinter.Label(plan_info_frame, text="Plan - Consult")
plan_con_label.grid(row=2, column=1)
plan_con_entry = tkinter.Entry(plan_info_frame)
plan_con_entry.grid(row=3, column=1)

plan_inter_label = tkinter.Label(plan_info_frame, text="Plan - Interventions")
plan_inter_label.grid(row=4, column=0)
plan_inter_entry = tkinter.Entry(plan_info_frame)
plan_inter_entry.grid(row=5, column=0)
for widget in plan_info_frame.winfo_children():
    widget.grid_configure(padx=10, pady=5)
###########################################################################
found_info_frame =tkinter.LabelFrame(frame, text="Foundations")
found_info_frame.grid(row= 5, column=0, padx=20, pady=10)

found_label = tkinter.Label(found_info_frame, text="Foundations")
found_label.grid(row=2, column=0)
found_entry = tkinter.Entry(found_info_frame)
found_entry.grid(row=3, column=0)

for widget in found_info_frame.winfo_children():
    widget.grid_configure(padx=10, pady=5)
##########################################################################


# Accept terms
terms_frame = tkinter.LabelFrame(frame, text="Review")
terms_frame.grid(row=6, column=0, sticky="news", padx=20, pady=10)

accept_var = tkinter.StringVar(value="Not Accepted")
terms_check = tkinter.Checkbutton(terms_frame, text= "I have reviewed all textual fields.",
                                  variable=accept_var, onvalue="Accepted", offvalue="Not Accepted")
terms_check.grid(row=0, column=0)

# Button
button = tkinter.Button(frame, text="Enter data", command= enter_data)
button.grid(row=7, column=0, sticky="news", padx=20, pady=10)



window.mainloop()
```

```python
#Standard Imports
import numpy as np
import re
import string
import pandas as pd
import random
```

```python
#load data
#Data nova_2 is actual text output from Deepgram Nova-2 merged with the file name,
#Dummy columns include student name, ID, email, and case (merged from SimulationIQ data)
```

```python
data_simIQ=pd.read_csv("AI Classification Data Formatted.csv", usecols=['ID','Organization Unit','Case/Scenario Name'],header=0)
data_simIQ= data_simIQ.sample(154).reset_index(drop=True) #randomly select 154 students
nova_2 = pd.read_csv("nova 2 transcriptions Final.csv", header=None)
```

```python
data_simIQ
```

```python
column_to_move =data_simIQ.pop("ID")
data_simIQ.insert(1, "Student Name", column_to_move)
```

```python
data_simIQ['Presentation']=nova_2
data_simIQ['ID'] = random.sample(range(1, 1000), 154)
data_simIQ['Email'] = data_simIQ['Student Name'] + '@vcom.edu'
```

```python
data_simIQ['ICD10'] = data_simIQ['Case/Scenario Name'].str.rsplit(' ',1).str[-1]
data_simIQ['Case/Scenario Name']=data_simIQ['Case/Scenario Name'].str.rsplit(' ',1).str[0]
data_simIQ
```

```python
column_to_move =data_simIQ.pop("ID")
data_simIQ.insert(0, "ID", column_to_move)
column_to_move =data_simIQ.pop("Email")
data_simIQ.insert(3, "Email", column_to_move)
column_to_move =data_simIQ.pop("Organization Unit")
data_simIQ.insert(3, "Organization Unit", column_to_move)
column_to_move =data_simIQ.pop("ICD10")
data_simIQ.insert(5, "ICD10", column_to_move)
```

```python
data_simIQ.to_csv('Simulated Data.csv',index=False)
```
